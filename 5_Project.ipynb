```python
import pandas as pd 
import numpy as np
from  sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

df = pd.read_csv("5_Student.csv")

x = df[['Hours']] # Always take 2D input [[]]
y = df['Score'] # Target coloum

model = LinearRegression()

model.fit(x, y)

predicted_score = model.predict(x)

mae = mean_absolute_error(y, predicted_score)
mse = mean_squared_error(y, predicted_score)
rmse = np.sqrt(mse)

print(f"MAE = {mae}")
print(f"MSE = {mse}")
print(f"RMSE = {rmse}")

new_prediction = model.predict([[7]])
print(f"predicted score for 7 hours is {new_prediction}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt 

df = pd.read_csv("Student_Performance.csv")

le = LabelEncoder()
df['school_type_encoded'] = le.fit_transform(df['school_type']) #public = 1, private = 0
df['internet_access_encoded'] = le.fit_transform(df['internet_access']) #Yes = 1, No = 0
df['extra_activities_encoded'] = le.fit_transform(df['extra_activities']) #Yes = 1, No = 0
df = pd.get_dummies(df, columns=['study_method'])
df = df.replace({True: 1, False: 0})

df = df.drop(columns= ['student_id', 'school_type', 'internet_access', 'extra_activities', 'age', 'gender', 'parent_education',
                        'travel_time', 'final_grade', 'math_score', 'science_score', 'english_score'])

# print(df.head())

minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df)

# print(pd.DataFrame(minmax_scaled, columns= df.columns)) # df.columns-select all columns


x = df.drop(columns= ['overall_score'])
y = df['overall_score']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# print("\nTraining Data")
# print(x_train)

# print("Testing Data")
# print(x_test)

# print("\nTraining Data")
# print(y_train)

# print("Testing Data")
# print(y_test)

model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred) # Closer to 1 = better`    `

print(f"MAE = {round(mae, 2)}")
print(f"MSE = {round(mse, 2)}")
print(f"RMSE = {round(rmse, 2)}")
print(f"R^2 score (model accuracy) = {round(r2, 2)}")


import joblib

joblib.dump(model, "student_model.pkl")

'''NEW PREDICTION '''

# study_hours = float(input("How many hours you studied : "))

# attendence_percentage = float(input("Attendence percentage : "))

# school_type = input("School type (public/private) : ").strip().lower()
# school_type_encoded = 1 if school_type == 'public' else 0

# internet_access = input("Internet access (yes/No) : ").strip().lower()
# internet_access_encoded = 1 if internet_access == 'yes' else 0

# extra_activities = input("You participate in extra activities (yes/No) : ").strip().lower()
# extra_activities_encoded = 1 if extra_activities == 'yes' else 0

# study_method = input("Your Study method (notes/textbook/group study/coaching/mixed/online videos) : ").strip().lower()
# study_method_coaching = 1 if study_method == 'coaching' else 0
# study_method_group_study = 1 if study_method == 'group_study' else 0
# study_method_mixed = 1 if study_method == 'mixed' else 0
# study_method_notes = 1 if study_method == 'notes' else 0
# study_method_online_videos = 1 if study_method == 'online_videos' else 0
# study_method_textbook = 1 if study_method == 'textbook' else 0


# y_pred = model.predict([[
#     study_hours,
#     attendence_percentage,
#     school_type_encoded,
#     internet_access_encoded,
#     extra_activities_encoded,
#     study_method_coaching,
#     study_method_group_study,
#     study_method_mixed,
#     study_method_notes,
#     study_method_online_videos,
#     study_method_textbook
# ]])

# print(f"Your overall score is {y_pred}")


# plt.Figure(figsize=(10, 6))
# plt.hist(df['overall_score'], bins= 30, color= 'pink', edgecolor='black')
# plt.title("Distribution of Overall_score")
# plt.xlabel("Final exam score")
# plt.ylabel("No. of students")
# plt.grid(True)
# plt.show()
```
